```python
from biopandas.pdb import PandasPdb
import pandas as pd
import numpy as np

ppdb = PandasPdb().read_pdb('HTR1F_19_364_4djh.pdb')

```

```python
sequence = ppdb.amino3to1()
sequence
```

```python
df = pd.DataFrame(ppdb.df['ATOM'])
```

```python
df
```

```python
xdf = df.groupby('residue_number', as_index=False)['x_coord'].mean()
ydf = df.groupby('residue_number', as_index=False)['y_coord'].mean()
zdf = df.groupby('residue_number', as_index=False)['z_coord'].mean()
```

```python
df = df.join(xdf.set_index('residue_number'), on='residue_number', how='right', lsuffix='_left', rsuffix='_right')
df = df.join(ydf.set_index('residue_number'), on='residue_number', how='right', lsuffix='_left', rsuffix='_right')
df = df.join(zdf.set_index('residue_number'), on='residue_number', how='right', lsuffix='_left', rsuffix='_right')
```

```python
df
```

```python
df = df[['residue_name', 'residue_number', 'x_coord_right', 'y_coord_right', 'z_coord_right']]
```

```python
df = df.drop_duplicates(subset='residue_number', keep="first")
df
```

```python
df['residue_name_convert'] = list(sequence['residue_name'])
df
```

```python
data = pd.read_csv('HTR1F.T2D.txt', header = None, sep="\t")
data['residue_number'] = data[0].str.strip().str[3:].str[:-1]
data['residue_name_convert'] = data[0].str.strip().str[2:3]
data = data.rename(columns={1: "score", 2: "p-val"})
data = data.astype({'residue_number': 'int64'})
data.dtypes
```

```python
df_tot = pd.merge(df, data, on="residue_number")
```

```python
df_tot
```

```python
df_risk = df_tot[df_tot['score'] < 0]
df_prot = df_tot[df_tot['score'] > 0]
df_prot = df_prot.drop_duplicates(subset=['x_coord_right'], keep='first')
df_risk = df_risk.drop_duplicates(subset=['x_coord_right'], keep='first')
```

```python
df_risk
```

```python
prots = np.array(df_prot[df_prot.columns[2:5]])
risks = np.array(df_risk[df_risk.columns[2:5]])
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dists_prots = euclidean_distances(prots, prots)
dists_risks = euclidean_distances(risks, risks)
```

```python
X = dists_prots

X = np.ma.masked_equal(X,0)

#other functionalities of masked arrays
X = X.compressed().reshape(24,23)

prots_vec = np.mean(X, axis=1)
emp_prot = np.mean(prots_vec)

Y = dists_risks

Y = np.ma.masked_equal(Y,0)
#other functionalities of masked arrays
Y = Y.compressed().reshape(17,16)

risks_vec = np.mean(Y, axis=1)
emp_risk = np.mean(risks_vec)
```

```python
from scipy.stats import mannwhitneyu

mannwhitneyu(risks_vec, prots_vec)
```

```python
dist_prots = []
dist_risks = []

x = np.array(df[df.columns[2:5]])
for i in range(100000):
    np.random.shuffle(x)
    
    prots = x[0:25,:]
    prots = np.unique(prots, axis=0)
    
    dists_prots = euclidean_distances(prots, prots)
    prots_vec = np.mean(dists_prots, axis=1)
    
    Y = dists_prots
    Y = np.ma.masked_equal(Y,0)
    nrow = Y.shape[0]
    
    #other functionalities of masked arrays
    Y = Y.compressed().reshape(nrow,nrow-1)
    print(np.mean(Y))
    dist_prots.append(np.mean(Y))
    
    
    risks = x[25:25+18,:]
    risks = np.unique(risks, axis=0)
    
    dists_risks = euclidean_distances(risks, risks)
    risks_vec = np.mean(dists_risks, axis=1)
    
    X = dists_risks
    X = np.ma.masked_equal(X,0)
    nrow = X.shape[0]
    
    #other functionalities of masked arrays
    X = X.compressed().reshape(nrow,nrow-1)
    dist_risks.append(np.mean(X))
    
```
